```python
import json
def printJson(ky):
    print(json.dumps(ky,ensure_ascii=False, encoding='UTF-8'))
   
```

```python
import os

'''
    读取目录下的所有文件并合并成一个内容块返回
'''

def readContent(filePath):
    _content=''
    _files=os.listdir(filePath)
    
    #help(os.path)
    
    for _f in _files:
        _fileFullPath=os.path.join(filePath,_f)
        _isFile=os.path.isfile(_fileFullPath)
        _isTxt=os.path.splitext(_fileFullPath)[1]=='.txt'
        
        #print((_isFile and _isTxt))
        
        #在python中是没有&&及||这两个运算符的，取而代之的是英文and和or。
        
        if (_isFile and _isTxt):
            print('---------loading---'+_fileFullPath+'-----\n')
            
            _content+=open(_fileFullPath,'r').read()
            _content+='\n'
    
    print('---------loaded--------\n')
    
    return _content
```

```python
import jieba.analyse as ja

# 这里使用jieba的textrank提取出_topK个关键词及其比重,默认参数的写法如下 

def getKeywords(_txts,_topK=1000):
    
    _emotion=('Ag', 'a', 'e', 'an','o','v','z','y')
    
    _n=('ns', 'n', 'vn', 'v')
    
    _res = ja.textrank(_txts, topK=_topK, withWeight=True,allowPOS=_emotion)
    
    
    #print(_res)

    # 生成关键词比重字典
    _keywords = dict()
    for i in _res:
        _k=i[0]
        _keywords[_k] = i[1]

    printJson(_keywords)
    
    return _keywords
```

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator

def generatorWordCloudImg(_keywords,_isImgColor,_model,_isBGW):
    
    _modelImg='./model/'+str(_model)+'.png'
    _font='./font/SourceHanSansCN-Light.otf'
    
    _max_words=800
    
    if _isBGW:
        _background_color='white'
    else:
        _background_color='black'
        
    
    _img = Image.open(_modelImg)
    #print(_img)
    _graph = np.array(_img)
    #print(_graph)
    
    _wc = WordCloud(font_path=_font,
    background_color=_background_color, max_words=_max_words, mask=_graph)

    #print(_wc)
    _wc.generate_from_frequencies(_keywords)
    _imgColor = ImageColorGenerator(_graph)

    # 保存词云图片
    if _isImgColor:
        _wc.recolor(color_func=_imgColor)
    
    _wc.to_file('./result/'+str(_model)+'_res.png')  
        
    
    #print(help(_wc))


```

```python
_txts=readContent('./data')
print(_txts[:12])
#print(_txts)

_keywords=getKeywords(_txts)

```
