```python
import requests
def page(word):
    url = 'http://www.baidu.com/s?wd='
    r=requests.get(url+word)
    #print r
    if r.status_code ==200:
        return r.text
    else:
        return False
```

```python
def init(html):#初始化
    soup = BeautifulSoup(html)
    results = []
    
   # print soup
    result_set = soup.find(id='content_left')
   # print result_set
    result_set = result_set.find_all('div',class_='c-container')
    
    for i in range(len(result_set)):
        result = result_set[i]
        
        a=__get_abstract(result)
        
        results.append(a)
        
    return results


# 获取abstract
def __get_abstract(result_div):
    if 'result-op' not in result_div['class']:  # 不是软广
        r_from = result_div.find(class_='c-abstract')
        if not r_from:
            return u'-'
        for em in r_from.find_all('em'):  # 移除abstract中的em标签
            em.unwrap()
        #print 0
       # print r_from.get_text()
        return r_from.get_text()
    else:
        return u'-'  
```

```python
def search(k):
    _page= page(k)
    results = init(_page)
    return results
```

```python
def guessAnswer(question,answers):
    _res=search(question)
    _strs=''.join(_res)
    _guess={}
    for ans in answers:
        _guess[ans]=_strs.count(ans)
    
    return _guess
```

```python
question=u'下列哪个戏种被称为中国的百戏之祖'
answers=[u'黄梅戏',u'越剧',u'昆曲']
u=guessAnswer(question,answers)
printJson(u)
```
