```python
from bs4 import BeautifulSoup
import requests
#_url = "http://image.baidu.com/search/index?tn=baiduimage&ps=1&ct=201326592&lm=-1&cl=2&nc=1&ie=utf-8&word=%E5%87%A4%E5%A7%90"
def page(word):
    _url = "http://www.baidu.com/s?wd="
    x = requests.get(_url+word)
    
  
    if x.status_code == 200:
        return x.text
    else:
        print(x.status_code)
        return False


  


def init(html):
    
    soup = BeautifulSoup(html)
    results = []
    
    result_set = soup.find(id="content_left")
    result_set = result_set.find_all("div",class_="c-container")
    
    for i in range(len(result_set)):
        result = result_set[i]
        mx = __get_abstract(result)
        results.append(mx)
        
       
    return results
        
# 获取abstract
def __get_abstract(result_div):
    if 'result-op' not in result_div['class']:  # 不是软广
        r_from = result_div.find(class_='c-abstract')
        if not r_from:
            return u'-'
        for em in r_from.find_all('em'):  # 移除abstract中的em标签
            em.unwrap()
        #print 0
       # print r_from.get_text()
        return r_from.get_text()
    else:
        return u'-'  

def search(keyword):
    _page = page(keyword)
    print _page
    results = init(_page)
    return results



```

```python
def guessanswer(question,answers):
    _res=search(question)
    _strs=''.join(_res)
    _guess={}
    for ans in answers:
       # print _strs
        _guess[ans]=_strs.count(ans)
         
    return _guess

```
